```python
import pandas as pd #pandas library
from pandas_datareader import data #data readers (google, html, etc.)
#The following line ensures that graphs are rendered in the notebook
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt #Plotting library
import datetime as dt #datetime for timeseries support
import os

# read, clean, filter data from textfile
# return df
def get_df(path):
    column_names = ["company", "location", "title", "views", "deadline", "published", "url"]
    df = pd.read_table(path, sep='\t', names=column_names, index_col=False, 
                            parse_dates=['deadline', 'published'])
    df = df[(df.url.str.startswith('http://www.cv.lt'))]
    df = df.drop_duplicates()
    return df


#saves chart as png file
#input:
#     series: series name
#     chart_kind: <string>
#     chart_title: <string>
#     saved_file_name: <string>, e.g. 'figure1.png'
def save_chart(series, chart_kind, chart_title, saved_file_name):
    import matplotlib.pyplot as plt #Plotting library
    import datetime as dt #datetime for timeseries support
    import os
    
    ax = series.plot(kind=chart_kind, title=chart_title, alpha=0.4, color='blue')
    fig = ax.get_figure()
    fig.savefig(os.path.join('charts/' + saved_file_name), dpi=100, format='png', bbox_inches='tight')
    fig.clf()
    
    return None
```

```python
#read data into df
df = get_df('data/job_ads_9_14.txt')
#prepare series for charts
chart1_series = df['title'][(df['published'] > '2017-07-20')].groupby(df.published.dt.date).size()
chart2_series = df.groupby(df.company).agg('count').sort_values(['title'], ascending=False)['title'][:10].sort_values()
chart3_series = df.groupby(df.location).agg('count').sort_values(['title'], ascending=False)['title'][:10]
chart4_series = chart4_series = df.sort_values(['views'], ascending=False)[:10][['title', 'views']].set_index('title')['views'].sort_values()

#for review
# chart1_series
# chart2_series
# chart3_series
# chart4_series

# save_chart(chart1_series, 'bar', 'Skelbimų skaičius pg. dieną', 'figure1.png')
# save_chart(chart2_series, 'barh', 'Skelbimų skaičius pg. įmonę', 'figure2.png')
# save_chart(chart3_series, 'bar', 'Skelbimų skaičius pg. vietovę', 'figure3.png')
# save_chart(chart4_series, 'barh', 'Populiariausi skelbimai pg. peržiūras', 'figure4.png')
```
